**Q 1. Find the Price of Product with ID 25.**

**Q 3. Retrieve the Supplier name of the product with the highest Price.**

**Answer:-**

image.png

**Formula Use :** =VLOOKUP(I14,D2:F102,3,0)

          Where: I14 is the Lookup Value
                D2:F102 is the Table Array
                3 is the Column Index Number.

**Q 4. Find the Category of the product with the lowest Stock ?**

**Answer:-**

image.png

**Formula Use :** =XLOOKUP(I19,E3:E102,C3:C102)

          Where: I19 is the LookUp Value.
                E3:E102 is the Lookup Array Range.
                C3:C102 is the Return Array Range.

**Q 5. Check if "Supplier_C" supplies any product in the "Electronics" category ?**

**Answer:-**

image.png

**Formula Use :** =COUNTIFS(C2:C102, "Electronics", F2:F102, "Supplier_C")

          Where: C2:C102 is the 1st Range .
                  "Electronics" is the Criteria 1.
                  F2:F102 is the 2nd Range.
                  "Supplier_C" is the Criteria 2.

**Q 6. Check if there is any product with a stock less than 100. If yes, find its name ?
(Hint : Use IF and VLOOKUP)**

**Answer:-**

image.png

**Formula Use :** =IF(J29<=100,VLOOKUP(A3,A2:F102,2,FALSE),"No Product")

          Where: J29<=100 is the Criteria.
                VLOOKUP(A3,A2:F102,2,FALSE) it is the True condition For  IF Statement.
                "No Product" it is the False Condition for the IF Statement.
